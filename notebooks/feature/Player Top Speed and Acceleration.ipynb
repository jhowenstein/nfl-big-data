```python
plays = analysis.return_plays()
```

```python
results = {}
for play in plays:
    
    players = play.return_defensive_players()
    
    for player in players:
        
        player_key = player.nflId
        
        if player_key not in results:
            results[player_key] = {}
            results[player_key]['max speed'] = [player.max_speed]
            results[player_key]['max accel'] = [player.max_accel]
        else:
            results[player_key]['max speed'].append(player.max_speed)
            results[player_key]['max accel'].append(player.max_accel)
```

```python
max_vals = {}
for key in results:
    max_vals[key] = [max(results[key]['max speed']),max(results[key]['max accel'])]
```

```python
df = pd.DataFrame.from_dict(max_vals,orient='index',columns=['max speed','max accel'])
```
