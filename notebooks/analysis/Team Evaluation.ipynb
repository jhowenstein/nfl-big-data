```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px

import os
import glob
```

```python
plt.rcParams['font.size'] = 14
```

```python
def plot_scheme(df, team, output=False, output_folder=''):
    team_data = df.loc[team]
    
    pie_names = ['Cover 0','Cover 1','Cover 2','Cover 3','Cover 4','Cover 6']
    pie_data_keys = ['cover 0 count','cover 1 count','cover 2 count','cover 3 count','cover 4 count','cover 6 count']
    pie_chart_data = team_data[pie_data_keys].values
    
    line_names = ['Overall','Cover 0','Cover 1','Cover 2','Cover 3','Cover 4','Cover 6']
    line_data_keys = ['overall epa rate','cover 0 epa rate','cover 1 epa rate','cover 2 epa rate','cover 3 epa rate','cover 4 epa rate','cover 6 epa rate']
    line_chart_data = team_data[line_data_keys].values
    
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(18,8))
    
    ax1.set_title('Coverage Scheme Percentages',fontsize=18)
    
    ax1.pie(pie_chart_data, labels=pie_names, autopct='%1.1f%%')
    
    ax2.set_title('Expected Points Added per Play',fontsize=18)
    ax2.set_ylabel('EPA / Play',fontsize=18)
    
    ax2.plot(line_names,line_chart_data,marker='o')
    ax2.set_xticklabels(line_names,rotation=-30)
    #ax2.set_ylim(-.75,.5)
    ax2.grid()
    
    plt.suptitle(team,fontsize=24)
    
    if output:
        plt.savefig(os.path.join(output_folder,f'{team}-SchemeViz.png'),dpi=100)
        
    plt.show()
```

```python
def plot_personnel(df, team, output=False, output_folder=''):
    team_data = df.loc[team]
    
    pie_names = ['Heavy (<4 DBs)','Standard','Nickel','Dime','Quarter']
    pie_data_keys = ['heavy count', 'standard count', 'nickel count','dime count', 'quarter count']
    pie_chart_data = team_data[pie_data_keys].values
    
    line_names = ['Overall', 'Heavy','Standard','Nickel','Dime','Quarter']
    line_data_keys = ['overall epa rate','heavy epa rate', 'standard epa rate','nickel epa rate', 'dime epa rate', 'quarter epa rate']
    line_chart_data = team_data[line_data_keys].values
    
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(18,8))
    
    ax1.set_title('Defensive Personnel Percentages',fontsize=18)
    
    ax1.pie(pie_chart_data, labels=pie_names, autopct='%1.1f%%',startangle=-60)
    
    ax2.set_title('Expected Points Added per Play',fontsize=18)
    ax2.set_ylabel('EPA / Play',fontsize=18)
    
    ax2.plot(line_names,line_chart_data,marker='o')
    ax2.set_xticklabels(line_names,rotation=-30)
    #ax2.set_ylim(-.25,.75)
    ax2.grid()
    
    plt.suptitle(team,fontsize=24)
    
    if output:
        plt.savefig(os.path.join(output_folder,f'{team}-PersonnelViz.png'),dpi=100)
        
    plt.show()
```

```python
def plot_target_coverage(df, team, output=False, output_folder=''):
    team_data = df.loc[team]
    
    pie_names = ['Man Coverage','Zone Coverage','Man+Zone Coverage','Checkdown','No Coverage']
    pie_data_keys = ['man count', 'zone count', 'man+zone count','checkdown count', 'no coverage count']
    pie_chart_data = team_data[pie_data_keys].values
    
    line_names = ['Overall','Man','Zone','Man+Zone','Checkdown','No Coverage']
    line_data_keys = ['overall epa rate','man epa rate','zone epa rate','man+zone epa rate','checkdown epa rate','no coverage epa rate']
    line_chart_data = team_data[line_data_keys].values
    
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(18,8))
    
    ax1.set_title('Targeted Defender Coverage Percentages',fontsize=18)
    
    ax1.pie(pie_chart_data, labels=pie_names, autopct='%1.1f%%')
    
    ax2.set_title('Expected Points Added per Play',fontsize=18)
    ax2.set_ylabel('EPA / Play',fontsize=18)
    
    ax2.plot(line_names,line_chart_data,marker='o')
    ax2.set_xticklabels(line_names,rotation=-30)
    #ax2.set_ylim(-.25,.75)
    ax2.grid()
    
    plt.suptitle(team,fontsize=24)
    
    if output:
        plt.savefig(os.path.join(output_folder,f'{team}-TargetCoverageViz.png'),dpi=100)
        
    plt.show()
```
