### Performance Analysis

```python
df = pd.merge(player_season_coverages[['displayName','position','snaps','man-total','zone-total']],
              man_production_reduction.drop(columns=['name','safety help percent']),
              left_index=True,right_index=True)
```

```python
df = pd.merge(df,zone_production_reduction.drop(columns=['name']),
              left_index=True,right_index=True)
```

```python
df
```

```python
snap_threshold = 200
```

```python
df = df[df['snaps'] > snap_threshold]
df
```

```python
df.keys()
```

```python
df['total targets'] = df['man targeted coverage count'] + df['zone targeted coverage count']
df['total delta epa'] = df['total man delta epa'] + df['total zone delta epa']
df['mean total delta epa'] = df['total delta epa'] / df['total targets']
```

```python
df['target %'] = ((df['total targets'] / df['snaps']) * 100).round(1)
df['man target %'] = ((df['man targeted coverage count'] / df['man-total']) * 100).round(1)
df['zone target %'] = ((df['zone targeted coverage count'] / df['zone-total']) * 100).round(1)
```

```python
df
```

### Limit Analysis to CB's

```python
cb_df = df[df['position'] == 'CB']
```

```python
def score(df, key, score_name, negative=True):
    mu = df[key].mean()
    s = df[key].std()
    
    if negative:
        df[score_name] = ((-(df[key] - mu) / s) * 10 + 50).round(1)
    else:
        df[score_name] = (((df[key] - mu) / s) * 10 + 50).round(1)
```

```python
score(cb_df, 'mean total delta epa', 'Total Coverage Score')
score(cb_df, 'mean man delta epa', 'Man Coverage Score')
score(cb_df, 'mean zone delta epa', 'Zone Coverage Score')
```

```python
score(cb_df, 'target %', 'Target % Score')
score(cb_df, 'man target %', 'Man Target % Score')
score(cb_df, 'zone target %', 'Zone Target % Score')
```

```python
sort_key = 'Total Coverage Score'
cb_df.sort_values(sort_key, ascending=False).head(10)
```

```python
display_keys = ['displayName','position','snaps','Total Coverage Score','Man Coverage Score','Zone Coverage Score',
                'Target % Score','Man Target % Score','Zone Target % Score']
display_cb_df = cb_df[display_keys].sort_values('Total Coverage Score',ascending=False)
```
