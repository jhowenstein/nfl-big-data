```python
import numpy as np
import pandas as pd
import scipy as sp

import matplotlib.pyplot as plt

import os
import glob
```

```python
for i in range(17):
    name = f'week{i+1}'
    print(name)
    df = pd.read_csv(os.path.join('data',name+'.csv'))
    
    N = df.shape[0]
    for k in range(df.shape[0]):
        if (k % 100000) == 0:
            print(f'  {int(k/N*100)}% Complete')
        if df.loc[k,'playDirection'] == 'left':
            df.at[k,'x'] = 120 - df.at[k,'x']
            df.at[k,'y'] = (160/3) - df.at[k,'y']
    df.to_csv(os.path.join('data',name+'-processed.csv'))
```
